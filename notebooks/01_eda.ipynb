# 01 — Exploratory Data Analysis

Goal: understand the dataset structure, degree of class imbalance, and the distribution of each feature.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set_theme(style='whitegrid')
%matplotlib inline

REPORT_DIR = '../reports/01_eda'
os.makedirs(REPORT_DIR, exist_ok=True)
```

## 1. Load Data

```python
df = pd.read_csv('../data/raw/creditcard.csv')
print(df.shape)
df.head()
```

```python
df.info()
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

## 2. Class Distribution (Imbalance)

```python
class_counts = df['Class'].value_counts()
print(class_counts)
print(f"Fraud rate: {class_counts[1] / len(df) * 100:.4f}%")

class_counts.plot(kind='bar', color=['steelblue', 'tomato'])
plt.title('Class Distribution')
plt.xticks([0, 1], ['Legit (0)', 'Fraud (1)'], rotation=0)
plt.ylabel('Count')
plt.tight_layout()
plt.savefig(f'{REPORT_DIR}/class_distribution.png', dpi=150)
plt.show()
```

## 3. Amount and Time Analysis by Class

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

for col_idx, (cls, label, color) in enumerate([(0, 'Legit', 'steelblue'), (1, 'Fraud', 'tomato')]):
    axes[0, col_idx].hist(df[df['Class'] == cls]['Amount'], bins=50, color=color)
    axes[0, col_idx].set_title(f'Amount — {label}')
    axes[0, col_idx].set_xlabel('Amount')
    axes[0, col_idx].set_ylabel('Count')

    axes[1, col_idx].hist(df[df['Class'] == cls]['Time'], bins=50, color=color)
    axes[1, col_idx].set_title(f'Time — {label}')
    axes[1, col_idx].set_xlabel('Time (seconds)')
    axes[1, col_idx].set_ylabel('Count')

plt.tight_layout()
plt.savefig(f'{REPORT_DIR}/amount_time_dist.png', dpi=150)
plt.show()
```

## 4. Feature Distributions by Class (All 30 Features)

```python
all_features = [f'V{i}' for i in range(1, 29)] + ['Amount', 'Time']

fig, axes = plt.subplots(6, 5, figsize=(22, 24))
axes = axes.flatten()

for i, col in enumerate(all_features):
    for cls, label, color in [(0, 'Legit', 'steelblue'), (1, 'Fraud', 'tomato')]:
        axes[i].hist(df[df['Class'] == cls][col], bins=40, alpha=0.5, label=label, color=color, density=True)
    axes[i].set_title(col)
    axes[i].legend(fontsize=7)

plt.suptitle('Feature Distributions by Class (All 30 Features)', y=1.01, fontsize=14)
plt.tight_layout()
plt.savefig(f'{REPORT_DIR}/feature_distributions_all.png', dpi=100, bbox_inches='tight')
plt.show()
```
