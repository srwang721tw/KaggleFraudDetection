# 01 — Exploratory Data Analysis

目標：了解資料結構、class imbalance 程度、各特徵分布與相關性。

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')
%matplotlib inline
```

## 1. 載入資料

```python
df = pd.read_csv('../data/raw/creditcard.csv')
print(df.shape)
df.head()
```

```python
df.info()
df.describe()
```

```python
# 缺失值檢查
df.isnull().sum()
```

## 2. Class 分布 (Imbalance)

```python
class_counts = df['Class'].value_counts()
print(class_counts)
print(f"詐欺比例: {class_counts[1] / len(df) * 100:.4f}%")

class_counts.plot(kind='bar', color=['steelblue', 'tomato'])
plt.title('Class Distribution')
plt.xticks([0, 1], ['Legit (0)', 'Fraud (1)'], rotation=0)
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('../reports/figures/class_distribution.png', dpi=150)
plt.show()
```

## 3. Amount 與 Time 分析

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

for cls, label, color in [(0, 'Legit', 'steelblue'), (1, 'Fraud', 'tomato')]:
    axes[0].hist(df[df['Class'] == cls]['Amount'], bins=50, alpha=0.6, label=label, color=color)
axes[0].set_title('Transaction Amount by Class')
axes[0].set_xlabel('Amount')
axes[0].legend()

for cls, label, color in [(0, 'Legit', 'steelblue'), (1, 'Fraud', 'tomato')]:
    axes[1].hist(df[df['Class'] == cls]['Time'], bins=50, alpha=0.6, label=label, color=color)
axes[1].set_title('Transaction Time by Class')
axes[1].set_xlabel('Time (seconds)')
axes[1].legend()

plt.tight_layout()
plt.savefig('../reports/figures/amount_time_dist.png', dpi=150)
plt.show()
```

## 4. PCA 特徵分布 (V1–V28)

```python
v_features = [f'V{i}' for i in range(1, 29)]

fig, axes = plt.subplots(7, 4, figsize=(20, 28))
axes = axes.flatten()

for i, col in enumerate(v_features):
    for cls, label, color in [(0, 'Legit', 'steelblue'), (1, 'Fraud', 'tomato')]:
        axes[i].hist(df[df['Class'] == cls][col], bins=40, alpha=0.5, label=label, color=color, density=True)
    axes[i].set_title(col)
    axes[i].legend(fontsize=7)

plt.suptitle('PCA Feature Distributions by Class', y=1.01, fontsize=14)
plt.tight_layout()
plt.savefig('../reports/figures/pca_feature_distributions.png', dpi=100, bbox_inches='tight')
plt.show()
```

## 5. 相關性 Heatmap

```python
corr = df.corr()

plt.figure(figsize=(20, 16))
sns.heatmap(corr, cmap='coolwarm', center=0, linewidths=0.3)
plt.title('Feature Correlation Heatmap')
plt.tight_layout()
plt.savefig('../reports/figures/correlation_heatmap.png', dpi=100)
plt.show()
```
